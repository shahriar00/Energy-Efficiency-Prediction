```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
x = data.drop(['Y1','Y2'],axis = 1)
y1 = data['Y1']
y2 = data['Y2']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y1,test_size = 0.3)
x_train,x_test,y_train,y_test = train_test_split(x,y2,test_size = 0.3)
```

```python
model = RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
x_pred = model.predict(x_train)
```

```python
metrics.mean_absolute_error(x_pred,y_train)
```

```python
metrics.r2_score(x_pred,y_train)
```

```python
metrics.mean_squared_error(x_pred,y_train)
```

```python
modeldata = LinearRegression()
modeldata.fit(x_train,y_train)
```

```python
x_predic = modeldata.predict(x_train)
```

```python
metrics.mean_squared_error(x_predic,y_train)
```

```python
metrics.r2_score(x_predic,y_train)
```
